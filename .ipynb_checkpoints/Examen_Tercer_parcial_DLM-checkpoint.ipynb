 * Título,

# Objetivo 
Predecir el valor de la variable **BoxOffice_USD** a partir de variables numéricas independientes disponibles en el dataset, tales como:<br>
* Budget_USD (Presupuesto en dólares)
* Duration (Duración en minutos)
* Ratting (Calificiación Promedio"

* Critic_reviews (Números de reseñas de criticos)

```python
# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
# Lectura de datos
data = pd.read_csv("res/movie_genre_classification_final.csv")
# Para ver las dimensiones
data.shape
# Ver primeros registros para validar la correcta lectura
data.head()
```

```python
# Visualización del conjunto de datos
var_independiente = ['Budget_USD', 'Duration', 'Rating', 'Votes', 'Num_Awards', 'Critic_Reviews']
var_dependiente = 'BoxOffice_USD'

# Tomado de IA Gemini
# Crear los gráficos de dispersión
plt.figure(figsize=(15, 10))
for i, feature in enumerate(var_independiente):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data[feature], kde=False, bins=20)  # bins controla el número de barras
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')
    plt.title(f'Distribución de {feature}')
    plt.grid(True)
plt.tight_layout()
plt.show()

#Para BoxOffice_USD
plt.figure(figsize=(8, 6))
sns.histplot(data['BoxOffice_USD'], kde=False, bins=20)
plt.xlabel('BoxOffice_USD')
plt.ylabel('Frecuencia')
plt.title('Distribución de BoxOffice_USD')
plt.grid(True)
plt.show()
```

```python
# Modificación del conjunto de datos
var_independiente = ['Budget_USD', 'Duration', 'Rating', 'Votes', 'Num_Awards', 'Critic_Reviews']
var_dependiente = 'BoxOffice_USD'

# Reemplazo de valores null(NaN) (reemplazar con la mediana) -- Tomado de IA Gemini
for i in var_independiente:
    data[i] = data[i].fillna(data[i].median())
data[var_dependiente] = data[var_dependiente].fillna(data[var_dependiente].median())
```

```python
# Se divide el set de datos en entrenamiento(80%) y prueba(20%)
X = data[var_independiente]
y = data[var_dependiente]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(type(X))
print(type(y))
```

```python
#Construcción del modelo de regresión lineal
# Tomado Tomado de IA Gemini - transforma los datos para que cada característica tenga una media de 0 y una desviación estándar de 1.
# Esto es importante para mejorar la interpretabilidad de muchos algoritmos de aprendizaje automático con varias características. 
X_train_np = X_train.values
X_test_np = X_test.values
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_np)
X_test_sc = scaler.transform(X_test_np)

# Se crea y entrena al modelo
model = LinearRegression()
model.fit(X_train_sc, y_train)

# Impresión del punto de intersección y sus pendientes
print("Intersección:", model.intercept_)
print("Pendiente:", model.coef_)

# Predicción
y_pred = model.predict(X_test_sc)
print(y_pred)
```

```python
# Predicción de nuevos ejemplos

# Ejemplo 1
ejemplo1 = pd.DataFrame({
    'Budget_USD': [1000000],
    'Duration': [120],
    'Rating': [8.1],
    'Votes': [50000],
    'Num_Awards': [1],
    'Critic_reviews': [100]
})
ejemplo1_np = ejemplo1.values   # Tomado de IA Gemini
ejemplo1_prediccion = model.predict(ejemplo1_np)
print(f"Predicción para el ejemplo 1: BoxOffice_USD = {ejemplo1_prediccion}")

# Tomado de IA Gemini
sns.histplot(y_test, kde=True, label="Distribución BoxOffice_USD (Test)")
plt.axvline(ejemplo1_prediccion[0], color='red', linestyle='--', label="Predicción Ejemplo 1")
plt.xlabel("BoxOffice_USD")
plt.ylabel("Densidad")
plt.title("Predicción del Ejemplo 1 vs. Distribución de BoxOffice_USD (Test)")
plt.legend()
plt.show()
```

```python
# Ejemplo 2
ejemplo2 = pd.DataFrame({
    'Budget_USD': [0],
    'Duration': [0],
    'Rating': [0],
    'Votes': [0],
    'Num_Awards': [0],
    'Critic_reviews': [0]
})
ejemplo2_np = ejemplo2.values   # Tomado de IA Gemini
ejemplo2_prediccion = model.predict(ejemplo2_np)
print(f"Predicción para el ejemplo 2: BoxOffice_USD = {ejemplo2_prediccion}")

# Tomado de IA Gemini
sns.histplot(y_test, kde=True, label="Distribución BoxOffice_USD (Test)")
plt.axvline(ejemplo2_prediccion[0], color='red', linestyle='--', label="Predicción Ejemplo 2")
plt.xlabel("BoxOffice_USD")
plt.ylabel("Densidad")
plt.title("Predicción del Ejemplo 2 vs. Distribución de BoxOffice_USD (Test)")
plt.legend()
plt.show()
```
